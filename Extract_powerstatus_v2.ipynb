```python
import datetime
date = []
begin = datetime.date(2022,1,1)
end = datetime.date(2022,6,29)
d = begin
delta = datetime.timedelta(days=1)
while d <= end:
    print(d.strftime("%Y%m%d"))
    date.append(str(d.strftime("%Y%m%d")))
    d += delta
```

```python
from bs4 import BeautifulSoup
import requests
import csv
import bs4

def check_link(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        print('can not connect to web')

def get_contents(rs):
    ulist = []
    soup = BeautifulSoup(rs, 'lxml')
    trs = soup.find_all('tr')
    for tr in trs:
        ui = []
        for td in tr:
            ui.append(td.text)
        ulist.append(ui)
    #print(ulist)
    for i in range(len(ulist)):
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
    #print(ulist)
    return ulist
```

```python
Path = "/Users/shuo/Desktop/powerstatus/2022_powerstatus.csv"
with open(Path, 'w',newline='') as f:
    writer = csv.writer(f)
    for datenumber in date:
        print(datenumber)
        url = 'https://www.nrc.gov/reading-rm/doc-collections/event-status/reactor-status/2022/' + datenumber + 'ps.html'
        rs = check_link(url)
        if rs == None:
            continue
        urlist = get_contents(rs)
        for i in range(len(urlist)):
            if len(urlist[i]) == 13:
                if (urlist[i][1] == '\xa0'): 
                    urlist[i][1] = ''
                if (urlist[i][3] == '\xa0'): 
                    urlist[i][3] = ''
                if (urlist[i][5] == '\xa0'): 
                    urlist[i][5] = ''
                if (urlist[i][7] == '\xa0'): 
                    urlist[i][7] = ''
                if (urlist[i][9] == '\xa0'): 
                    urlist[i][9] = ''
                if (urlist[i][11] == '\xa0'): 
                    urlist[i][11] = ''
                writer.writerow([datenumber, urlist[i][1], urlist[i][3], urlist[i][5], urlist[i][7], urlist[i][9], urlist[i][11]])
            if len(urlist[i]) == 11:
                if (urlist[i][1] == '\xa0'): 
                    urlist[i][1] = ''
                if (urlist[i][3] == '\xa0'): 
                    urlist[i][3] = ''
                if (urlist[i][5] == '\xa0'): 
                    urlist[i][5] = ''
                if (urlist[i][7] == '\xa0'): 
                    urlist[i][7] = ''
                if (urlist[i][9] == '\xa0'): 
                    urlist[i][9] = ''
                writer.writerow([datenumber, urlist[i][1], urlist[i][3], urlist[i][5], urlist[i][7], urlist[i][9]])
            if len(urlist[i]) == 6:
                if (urlist[i][0] == '\xa0'): 
                    urlist[i][0] = ''
                if (urlist[i][1] == '\xa0'): 
                    urlist[i][1] = ''
                if (urlist[i][2] == '\xa0'): 
                    urlist[i][2] = ''
                if (urlist[i][3] == '\xa0'): 
                    urlist[i][3] = ''
                if (urlist[i][4] == '\xa0'): 
                    urlist[i][4] = ''
                if (urlist[i][5] == '\xa0'): 
                    urlist[i][5] = ''
                writer.writerow([datenumber, urlist[i][0], urlist[i][1], urlist[i][2], urlist[i][3], urlist[i][4], urlist[i][5]])
f.close
```
