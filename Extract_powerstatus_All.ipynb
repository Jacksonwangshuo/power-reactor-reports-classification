```python
import datetime
def date(i):

    date = []
    begin = datetime.date(int(f'{i}'),1,1)
    end = datetime.date(int(f'{i}'),12,31)
    d = begin
    delta = datetime.timedelta(days=1)
    while d <= end:
        #print(d.strftime("%Y%m%d"))
        date.append(str(d.strftime("%Y%m%d")))
        d += delta
    return date
```

```python
from bs4 import BeautifulSoup
import requests
import csv
import bs4
import datetime

def check_link(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        print('can not connect to web')

def get_contents(ulist, rurl):
    soup = BeautifulSoup(rurl, 'lxml')
    trs = soup.find_all('tr')
    for tr in trs:
        ui = []
        for td in tr:
            ui.append(td.text)
        ulist.append(ui)
    #print(ulist)
    for i in range(len(ulist)):
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
        if '\n' in ulist[i]:
            ulist[i].remove('\n')
    #print(ulist)
    return ulist
```

```python
urli = []
with open("/Users/shuo/Desktop/powerstatus.csv", 'w',newline='') as f:
    writer = csv.writer(f)
    for k in range(2010, 2022):
        print(k)
        for datenumber in date(k):
            print(datenumber)
            url = 'https://www.nrc.gov/reading-rm/doc-collections/event-status/reactor-status/' + str(k) + '/' + datenumber + 'ps.html'
            print(url)
            rs = check_link(url)
            urlist = get_contents(urli, rs)
            for i in range(len(urlist)):
                writer.writerow([datenumber, urlist[i][0], urlist[i][1], urlist[i][2], urlist[i][3], urlist[i][4], urlist[i][5]])
f.close
```
